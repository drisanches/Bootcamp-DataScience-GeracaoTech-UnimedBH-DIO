## **Estrutura de Dados em Python:**

*   **Listas:** utilizam colchetes [] como sintaxe e são mutáveis;
*   **Tuplas:** usam parênteses () como sintaxe e são imutáveis, ou seja, uma vez criadas não permitem a alteração de nenhum elemento;
*   **Dicionários:** usam chaves {} como sintaxe e trabalham com o conceito de chave e valor.





## **Manipulação de Dados com a Biblioteca Pandas**

```python
# Importando a Biblioteca Pandas
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/DataSets/Gapminder.csv", error_bad_lines=False, sep=";")
```

```python
# Visualizando as 5 primeiras linhas:
df.head()
```

```python
# Renomeando os títulos das colunas e incorporando ao dataframe
df = df.rename(columns={"country": "pais", "continent": "continente", "year": "ano", "lifeExp": "Expectativa de Vida", "pop": "Pop Total", "gdpPercap": "PIB"})
```

```python
# Retornar informações estatísticas do conjunto de dados
df.describe()
```

```python
# Exibir dados de apenas uma coluna sem repetição
df["continente"].unique()
```

```python
# Fazendo fitros (método loc): exibir dados somente de um continente
Oceania = df.loc[df["continente"] == "Oceania"]
Oceania.head()
```

### **Agrupamento de Dados:**

```python
# Quantos países existem em cada continente?
df.groupby("continente")["pais"].nunique()
```

```python
# Qual a expectativa média de vida de cada ano?
df.groupby("ano")["Expectativa de Vida"].mean()
```

```python
# Calcular a média geral
df["PIB"].mean()
```

```python
# Calcular a soma de determinada coluna
df["PIB"].sum()
```
